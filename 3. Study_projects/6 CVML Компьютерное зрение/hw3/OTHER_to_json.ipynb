Обработываем чужие ректенжелы

```python
train_files = glob.glob('/content/drive/MyDrive/fish/NoF/*')
```

Обработка областей разметки для файлов OTHER, найденной где-то в интернете

```python
to_json = []

for file in train_files:
  with open(file, "r") as f:
    rect = []
    for line_ in f:
      line = list(
          map(
              float, line_.split(",")
              )
          )
      data_dict = {'class': 'rect',
                   'height': line[2],
                   'width': line[3],
                   'x': line[1],
                   'y': line[0]}
      rect.append(data_dict)

    name = re.search("NoF\/(.*?)\.tsv2", file)[1]

    to_json.append(
        {'annotations': [rect_ for rect_ in rect], 
         'class': 'image',
         'filename': name}
        )
```

Создаём области для категории NoF - нет рыбы. Область будет представлять собой всю обюласть фотографии, за вычетом полей по бокам.

```python
import cv2

to_json = []

for file in tqdm(train_files):
    pic = cv2.imread(file)
    data_dict = {'class': 'rect',
                   'height': pic.shape[0] - 60,
                   'width': pic.shape[1] - 20,
                   'x': 10,
                   'y': 50}
    rect = []
    rect.append(data_dict)

    name = re.search("NoF\/(.*?\.jpg)", file)[1]

    to_json.append(
        {'annotations': [rect_ for rect_ in rect], 
         'class': 'image',
         'filename': name}
        )
```

```python
with open('nof_labels.json', 'w') as outfile:
    json.dump(to_json, outfile)
```
